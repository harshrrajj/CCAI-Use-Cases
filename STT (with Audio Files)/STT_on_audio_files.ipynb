```python
import torch
import librosa  # A python package for music and audio analysis.
import numpy as np
import soundfile as sf  # used to read and process audio files.
from scipy.io import wavfile # used to read and process audio files.
from IPython.display import Audio
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer 
from pydub import AudioSegment
import time
```

```python
# This cell downloads the model specified.
tokenizer = Wav2Vec2Tokenizer.from_pretrained("wav2vec2-base-timit-demo-colab") # The tokenizer class object.
model = Wav2Vec2ForCTC.from_pretrained("wav2vec2-base-timit-demo-colab")        # The predictor class object.
```

```python
audio_path = '/content/test-audio.wav' # add the audio file to be transcribed to a directory.
audio_path1 = '/content/test-audio-2.wav'  # add another audio file to be transcribed to a directory.
```

```python
def speechToText(pathToAudio, interval):    # streams and transcribes the audio file provided to it.

  inputAudio, _ = librosa.load(pathToAudio, sr = 16000)  #Librosa is used to convert the audio to its corresponding float array.
  # This float array will be further divided into chunks (below)
  end = 0
  
  while(end<len(inputAudio)):
    if((len(inputAudio))<=(end+interval)):
      end = len(inputAudio)
    else:
      end+=interval   # each chunk size = [interval]
      
    input_values = tokenizer(inputAudio[0:end], return_tensors="pt").input_values 
    logits = model(input_values).logits
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = tokenizer.batch_decode(predicted_ids)[0]
    print(transcription)
```

```python
speechToText(audio_path, 4500)  #Note: length of inputAudio array for audio_path is 126559

# So, the time it takes to transcribe a streaming audio directly depends on the interval size(or chunk size) (here, it's 10000). This value needs
# to be adjusted according to need or according to the length of the audio. Larger the chunk size, lesser the time it takes to process.

''' One way to decide the value of the interval/chunk size can be:
    taking the ratio of size(inputAudio array) to size(audio file in seconds) because this way, we might get to see texts on the output screen almost ever second and the whole process will expedite.
    Further, for larger audio files or for an unknown audio length, we can even set a limit for the number of chunks or divisions of an audio, and the streaming can be done accordingly.
'''
```

```python
def returnChunkSize(pathToAudio):   # returns an ideal/optimal chunk size for a given audio for streaming
  newAudio = AudioSegment.from_wav(pathToAudio)
  audioLength = int(len(newAudio) / 1000.0)  # in seconds
  inputAudioArray, _ = librosa.load(pathToAudio, sr = 16000)
  arrayLength = len(inputAudioArray)
  chunkSize = int(arrayLength / audioLength)
  return chunkSize

speechToText(audio_path1, returnChunkSize(audio_path1))  # This line/cell can be run to test the same on a longer audio file (test-audio-2.wav). WORKS FINE!!
```

```python
import torch
import zipfile
import torchaudio
from glob import glob

device = torch.device('cpu')  # cpu-extensive
#Load the model and its utilities.
model, decoder, utils = torch.hub.load(repo_or_dir='snakers4/silero-models',
                                       model='silero_stt',
                                       language='en', # also available 'de', 'es'
                                       device=device)
(read_batch, split_into_batches,
 read_audio, prepare_model_input) = utils 

test_file = glob('/content/test-audio.wav')
test_file1 = glob('/content/test-audio-2.wav')
```

```python
def printSTTexample(testFile):
  """ This is an example of batch transcription """
  batches = split_into_batches(testFile, batch_size=10)
  input = prepare_model_input(read_batch(batches[0]),
                              device=device)

  output = model(input)
  for example in output:
      print(decoder(example.cpu()))
```

```python
from deepspeech import Model
import numpy as np
import os
import wave
from IPython.display import Audio
```

```python
model_file_path = 'deepspeech-0.8.2-models.pbmm'  #for acoustic
lm_file_path = 'deepspeech-0.8.2-models.scorer'   #for language

#Model parameters
beam_width = 100
lm_alpha = 0.93
lm_beta = 1.18

model = Model(model_file_path)
model.enableExternalScorer(lm_file_path)

model.setScorerAlphaBeta(lm_alpha, lm_beta)
model.setBeamWidth(beam_width)
```

```python
stream = model.createStream()   #its inbuilt function .createStream() facilitates audio streaming property.
```

```python
def read_audio_file(filename):

  with wave.open(filename, 'rb') as w:
    rate = w.getframerate()
    frames = w.getnframes()
    buffer = w.readframes(frames)

  return buffer, rate
```

```python
def real_time_transcription(audio_file):

  buffer, rate = read_audio_file(audio_file)
  offset=0
  batch_size=8196 #chunk size
  text=''

  while offset < len(buffer):
    end_offset=offset+batch_size
    chunk=buffer[offset:end_offset]
    data16 = np.frombuffer(chunk, dtype=np.int16)
    stream.feedAudioContent(data16)
    text=stream.intermediateDecode()
    print(text)
    offset = end_offset

  return True
```

# Detect Silent and Non-silent intervals in an audio file using Pydub


```python
!pip install pydub
```

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence, detect_nonsilent, detect_silence

audioFilePath = '/content/test-audioSil.wav'  # An audio file with some silence intervals
Audio(audioFilePath)
```

```python
# our target is to detect the silent and non-silent intervals in the above audio-file, which can be fulfilled by using the pydub package in python.

sound = AudioSegment.from_wav(audioFilePath)
print("Length of audio={} seconds".format(len(sound)/1000)) # overall length of the audio

silentData = detect_silence(sound, min_silence_len=1500, silence_thresh=sound.dBFS-16)  # the threshold value for silence is dbFS-16 (this is a standard value), and the speaker must be silent for at least 1.5 seconds.

# print the starting and stopping time stamps of the silent intervals.
print("Start,  Stop")
for chunks in silentData:
  print( [chunk/1000 for chunk in chunks])
```
